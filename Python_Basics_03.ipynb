# Important note about variable scope in Python

 - You can have local and global scopes in python. 
 - Python variables are scoped to the innermost function, class, or module in which they're assigned. 
 - Control blocks don't count, so a variable assigned inside an 'if' block is does not have its scope limited to that block.

```python
# Example

num1=60 
num2=40 
num3=20

if num1 >= num2 and num1 >= num3:
    largest = num1    
elif num2 >= num1 and num2 >= num3:
    largest= num2
else:
    largest= num3
print ("largest no among three is {}".format(largest))
```

# Python while loop

- used to iterate over a block of code ( inside the while loop body ) till the condition evaluted by the while loop is True. 

```python
# Example

# Getting the product of all the elements of a list

List = [1,2,3,4,5]

# Use a new variable to store the product vaule. 
# Use a initializer variable for the while loop to iterate over all the elements inside the while loop while multiplying

product = 1
index = 0

while index < len(List):                # Till value of index is less then the number of elements in the List
    product *= List[index]              # multiply the value inside vairable product with each element in the list 
                                        #and store the new value back into variable product
        
    index += 1                  # increse the value of the index by 1 after each test condition evaluation till it is true

print("The product of all the elements in the given list is : {}" .format(product))

```

# Example for while loop with else 

```python
List = [1,2,3,4,5]

index = 0

while index < len(List):
    print(List[index])
    index += 1

else:
    print("All numbers from the list are printed now. ")
```

# Program to check whether a number is prime or not

```python
# Method - check if the given number is divisible by any number less than square root of the number.
# if yes, then it is a composite number otherwise it is a prime number

import math

num = int(input("Please enter a number to check if it is prime or composite :"))

sq_num = math.sqrt(num)

diviser = 2
reminder = 0

while diviser <= sq_num :
    reminder = num % diviser
    if reminder == 0:
        print("The numbered entered is a composite number.")
        break
    else:    
        diviser += 1
else:
    print("The numbered enterted is a prime number")
    


```

# For Loop

- For loop is used to iterate over a sequence ( lists, tuples, sets, dictionaries ) or any other iteratable objects.
- Iterating over a sequence is called traversal.

```python
# Example 

# Getting the product os all the elements in a list using for loop

list = [2,3,5,6,4,7,8]
product = 1

for ele in list:
    product *= ele

print("The product of all the elements in the list is : {}".format(product))
```

# Range function

- We can use range() function to generate a sequence of numbers. 
- e.g. range(10) will generate all numbers from 0 to 9 (10 numbers).
- We can also efine the start, stop and step size as range(start,stop,step size). Default step size = 1.
- range funtion does not store all generated numbers in the memory. 


```python
# Example

for i in range(10):
    print(i) 
```

```python
for i in range(2,200,15):
    print(i)
```

```python
# iterate over a list using indix

First_Name = ["Karan", "Ram" , "Shyam" , "Nada"]

for index in range(len(First_Name)):
    print(First_Name[index])
```

```python
# iterate over a list without using range() function

First_Name = ["Karan", "Ram" , "Shyam" , "Nada"]

for ele in First_Name:
    print(ele)
```

# Python program to write down all the prime numbers in a given range

```python
start = int(input("enter the start point for the range to find the number of primes :"))
end = int(input("enter the end point for the range to find the number of primes :"))

value = start
isDivisible = False

for ele in range(start, end+1):
    for diviser in range(2,value):
        if ele % diviser == 0:
            isDivisible = True
            value += 1
        else:
            print(ele)   

print("All prime numbers between {} and {} has been shown above. " .format(start,end))
   
```

```python
import math

start = int(input("enter the start point for the range to find the number of primes :"))
end = int(input("enter the end point for the range to find the number of primes :"))

for ele in range(start, end+1):
    sqrt = math.floor(math.sqrt(ele))
    isDivisible = False;
    for num in range(2,sqrt+1):
        if ele % num == 0:
            isDivisible = True;
        if not isDivisible:
            print(ele);

print("All prime numbers between {} and {} has been shown above. " .format(start,end))

```
